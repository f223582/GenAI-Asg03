```python
# CYCLEGAN - STEP 1: DATA PREPARATION WITH SKETCHY DATASET
import os, random
import numpy as np
from PIL import Image
from pathlib import Path
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
import torchvision.utils as vutils
import matplotlib.pyplot as plt

# Seed
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
set_seed(42)

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
if torch.cuda.is_available():
    from torch.amp import GradScaler
    scaler = GradScaler('cuda')
else:
    scaler = None
print(f"PyTorch: {torch.__version__} | Device: {device} | GPUs: {torch.cuda.device_count()}")

# STEP 1: Sketchy dataset paths
SKETCHY_ROOT = Path("/kaggle/working/sketchy/256x256")

# Recommended augmentations (from README)
PHOTO_SUBDIR = "photo/tx_000100000000"      # Bounding box +10% (best for CycleGAN)
SKETCH_SUBDIR = "sketch/tx_000000000000"    # Matched with photo

PHOTO_DIR = SKETCHY_ROOT / PHOTO_SUBDIR
SKETCH_DIR = SKETCHY_ROOT / SKETCH_SUBDIR

print("="*60)
print("SKETCHY DATASET LOADING")
print("="*60)
print(f"📸 Photo directory: {PHOTO_DIR}")
print(f"✏️ Sketch directory: {SKETCH_DIR}")
print(f"Photo exists: {PHOTO_DIR.exists()}")
print(f"Sketch exists: {SKETCH_DIR.exists()}")

# Output directories for organized data
sketch_out = "/kaggle/working/sketchy_sketches"
photo_out = "/kaggle/working/sketchy_photos"
os.makedirs(sketch_out, exist_ok=True)
os.makedirs(photo_out, exist_ok=True)


# STEP 1B: COPY PHOTOS TO WORKING DIRECTORY

already_photos = list(Path(photo_out).rglob("*.jpg")) + list(Path(photo_out).rglob("*.png"))

if len(already_photos) >= 500:
    print(f"\n✅ Photos already ready: {len(already_photos)}")
else:
    print("\n📸 Copying photos from Sketchy dataset...")
    all_photos = list(PHOTO_DIR.rglob("*.jpg"))
    print(f"   Total photos found: {len(all_photos)}")
    
    random.shuffle(all_photos)
    count = 0
    for img_file in tqdm(all_photos[:2000], desc="Copying photos"):
        try:
            img = Image.open(img_file).convert("RGB")
            img.save(f"{photo_out}/{count:04d}.jpg")
            count += 1
        except Exception as e:
            continue
    
    print(f"✅ Photos copied: {count}")


# STEP 1C: COPY SKETCHES TO WORKING DIRECTORY

already_sketches = list(Path(sketch_out).rglob("*.png"))

if len(already_sketches) >= 500:
    print(f"\n✅ Sketches already ready: {len(already_sketches)}")
else:
    print("\n✏️ Copying sketches from Sketchy dataset")
    all_sketches = list(SKETCH_DIR.rglob("*.png"))
    print(f"   Total sketches found: {len(all_sketches)}")
    
    random.shuffle(all_sketches)
    count = 0
    for img_file in tqdm(all_sketches[:2000], desc="Copying sketches"):
        try:
            img = Image.open(img_file).convert("RGB")
            img.save(f"{sketch_out}/{count:04d}.png")
            count += 1
        except Exception as e:
            continue
    
    print(f"✅ Sketches copied: {count}")


# STEP 1D: VERIFY COUNTS
final_photos = len(list(Path(photo_out).rglob("*.jpg")))
final_sketches = len(list(Path(sketch_out).rglob("*.png")))

print("\n" + "="*60)
print("DATASET SUMMARY")
print("="*60)
print(f"📸 Total photos: {final_photos}")
print(f"✏️ Total sketches: {final_sketches}")
print(f"📊 Total images: {final_photos + final_sketches}")

# STEP 1E: CONFIGURATION


class Config:
    IMG_SIZE       = 128
    CHANNELS       = 3
    BATCH_SIZE     = 6
    NUM_EPOCHS     = 50
    LR             = 0.0002
    BETAS          = (0.5, 0.999)
    LAMBDA_CYC     = 10
    LAMBDA_ID      = 5
    N_RES_BLOCKS   = 6
    MAX_SAMPLES    = min(final_photos, final_sketches, 2000)
    SKETCH_DIR     = "/kaggle/working/sketchy_sketches"
    PHOTO_DIR      = "/kaggle/working/sketchy_photos"
    CHECKPOINT_DIR = "/kaggle/working/checkpoints"
    SAMPLE_DIR     = "/kaggle/working/samples"

cfg = Config()
os.makedirs(cfg.CHECKPOINT_DIR, exist_ok=True)
os.makedirs(cfg.SAMPLE_DIR, exist_ok=True)

print(f"\n⚙️ CONFIGURATION")
print("="*60)
print(f"   Image size: {cfg.IMG_SIZE}x{cfg.IMG_SIZE}")
print(f"   Batch size: {cfg.BATCH_SIZE}")
print(f"   Epochs: {cfg.NUM_EPOCHS}")
print(f"   Max samples: {cfg.MAX_SAMPLES}")
print(f"   Learning rate: {cfg.LR}")
print(f"   Cycle loss weight: {cfg.LAMBDA_CYC}")
print(f"   Identity loss weight: {cfg.LAMBDA_ID}")
print(f"   ResNet blocks: {cfg.N_RES_BLOCKS}")

# STEP 1F: TRANSFORMS + DATASET CLASS


transform = transforms.Compose([
    transforms.Resize((cfg.IMG_SIZE, cfg.IMG_SIZE)),
    transforms.RandomHorizontalFlip(p=0.5),  # Data augmentation
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

class DomainDataset(Dataset):
    def __init__(self, path, transform=None, max_samples=2000):
        self.transform = transform
        exts = {'.jpg', '.jpeg', '.png', '.bmp'}
        self.imgs = []
        for ext in exts:
            self.imgs.extend(list(Path(path).rglob(f"*{ext}")))
        random.shuffle(self.imgs)
        self.imgs = self.imgs[:max_samples]
        print(f"  📁 {len(self.imgs)} images from: {Path(path).name}")

    def __len__(self):
        return len(self.imgs)

    def __getitem__(self, idx):
        try:
            img = Image.open(self.imgs[idx]).convert("RGB")
        except Exception:
            img = Image.new("RGB", (cfg.IMG_SIZE, cfg.IMG_SIZE), 0)
        
        if self.transform:
            img = self.transform(img)
        return img


# STEP 1G: CREATE DATALOADERS


print("\n📦 Creating DataLoaders...")
print("="*60)

sketch_dataset = DomainDataset(cfg.SKETCH_DIR, transform, cfg.MAX_SAMPLES)
photo_dataset = DomainDataset(cfg.PHOTO_DIR, transform, cfg.MAX_SAMPLES)

sketch_loader = DataLoader(
    sketch_dataset, 
    batch_size=cfg.BATCH_SIZE,
    shuffle=True, 
    num_workers=2,
    pin_memory=True, 
    drop_last=True
)

photo_loader = DataLoader(
    photo_dataset, 
    batch_size=cfg.BATCH_SIZE,
    shuffle=True, 
    num_workers=2,
    pin_memory=True, 
    drop_last=True
)

print(f"\n✅ Sketch loader: {len(sketch_loader)} batches")
print(f"✅ Photo loader: {len(photo_loader)} batches")


# STEP 1H: VISUALIZE SAMPLES


def denorm(tensor):
    """Denormalize from [-1, 1] to [0, 1]"""
    return (tensor * 0.5) + 0.5

def show_samples(loader, title, n=6):
    """Display sample images from dataloader"""
    batch = next(iter(loader))[:n]
    grid = vutils.make_grid(denorm(batch), nrow=n, padding=2)
    
    plt.figure(figsize=(15, 3))
    plt.imshow(grid.permute(1, 2, 0).cpu().numpy())
    plt.title(title, fontsize=14, fontweight='bold')
    plt.axis("off")
    plt.tight_layout()
    
    # Save sample
    sample_path = f"{cfg.SAMPLE_DIR}/{title.replace(' ', '_').replace('-', '')}.png"
    plt.savefig(sample_path, dpi=100, bbox_inches='tight')
    print(f"   💾 Saved: {sample_path}")
    plt.show()

print("\n" + "="*60)
print("SAMPLE VISUALIZATION")
print("="*60)

show_samples(sketch_loader, "Domain A - Sketches (Sketchy Dataset)")
show_samples(photo_loader, "Domain B - Photos (Sketchy Dataset)")


# STEP 1I: SANITY CHECKS


print("\n" + "="*60)
print("SANITY CHECKS")
print("="*60)

# Get one batch from each loader
sketch_batch = next(iter(sketch_loader))
photo_batch = next(iter(photo_loader))

print(f"📊 Sketch batch shape: {sketch_batch.shape}")
print(f"📊 Photo batch shape: {photo_batch.shape}")
print(f"📈 Sketch value range: [{sketch_batch.min():.2f}, {sketch_batch.max():.2f}]")
print(f"📈 Photo value range: [{photo_batch.min():.2f}, {photo_batch.max():.2f}]")

# Assertions
assert sketch_batch.shape == (cfg.BATCH_SIZE, 3, cfg.IMG_SIZE, cfg.IMG_SIZE), \
    f"Sketch batch shape mismatch: {sketch_batch.shape}"
assert photo_batch.shape == (cfg.BATCH_SIZE, 3, cfg.IMG_SIZE, cfg.IMG_SIZE), \
    f"Photo batch shape mismatch: {photo_batch.shape}"
assert -1.1 <= sketch_batch.min().item() <= 0, \
    f"Sketch normalization issue: min={sketch_batch.min():.2f}"
assert -1.1 <= photo_batch.min().item() <= 0, \
    f"Photo normalization issue: min={photo_batch.min():.2f}"

print("\n✅ All sanity checks passed!")


# STEP 1J: DATASET INFORMATION

print("\n" + "="*60)
print("DATASET INFORMATION")
print("="*60)

# Category information
photo_categories = [d.name for d in Path(PHOTO_DIR).iterdir() if d.is_dir()]
sketch_categories = [d.name for d in Path(SKETCH_DIR).iterdir() if d.is_dir()]

print(f"📁 Photo categories: {len(photo_categories)}")
print(f"📁 Sketch categories: {len(sketch_categories)}")

# Show first 10 categories
print(f"\n📋 First 10 categories:")
for i, cat in enumerate(sorted(photo_categories)[:10]):
    print(f"   {i+1:2}. {cat}")

# Check if categories match
common_cats = set(photo_categories) & set(sketch_categories)
print(f"\n🔄 Common categories: {len(common_cats)}")

# Training information
total_steps_per_epoch = min(len(sketch_loader), len(photo_loader))
total_steps = total_steps_per_epoch * cfg.NUM_EPOCHS

print(f"\n📊 TRAINING STATISTICS:")
print(f"   Steps per epoch: {total_steps_per_epoch}")
print(f"   Total steps: {total_steps}")
print(f"   Estimated time: ~{total_steps_per_epoch * cfg.NUM_EPOCHS * 0.03 / 60:.1f} hours")
print(f"   (Assuming ~0.03 sec/step on T4 GPU)")


# STEP 1K: SAVE CONFIGURATION

import json

config_info = {
    "dataset": "Sketchy Dataset",
    "photo_dir": str(PHOTO_DIR),
    "sketch_dir": str(SKETCH_DIR),
    "num_photos": final_photos,
    "num_sketches": final_sketches,
    "num_categories": len(photo_categories),
    "common_categories": len(common_cats),
    "image_size": cfg.IMG_SIZE,
    "batch_size": cfg.BATCH_SIZE,
    "num_epochs": cfg.NUM_EPOCHS,
    "learning_rate": cfg.LR,
    "lambda_cycle": cfg.LAMBDA_CYC,
    "lambda_identity": cfg.LAMBDA_ID,
    "n_res_blocks": cfg.N_RES_BLOCKS,
    "max_samples": cfg.MAX_SAMPLES
}

with open(f"{cfg.CHECKPOINT_DIR}/config.json", "w") as f:
    json.dump(config_info, f, indent=2)

print(f"\n💾 Configuration saved to: {cfg.CHECKPOINT_DIR}/config.json")


# FINAL SUMMARY

print("\n" + "="*60)
print("🎉 STEP 1 COMPLETE! DATA PREPARATION DONE!")
print("="*60)
print(f"""
✅ Dataset: Sketchy Dataset
✅ Photos loaded: {final_photos}
✅ Sketches loaded: {final_sketches}
✅ Categories: {len(common_cats)} common categories
✅ DataLoaders ready: {len(sketch_loader)} sketch batches, {len(photo_loader)} photo batches
✅ All sanity checks passed
✅ Ready for CycleGAN training!

📁 Output directories:
   - Sketches: {cfg.SKETCH_DIR}
   - Photos: {cfg.PHOTO_DIR}
   - Checkpoints: {cfg.CHECKPOINT_DIR}
   - Samples: {cfg.SAMPLE_DIR}

🚀 Next step: Build Generator and Discriminator networks!
""")
```

```python

# CYCLEGAN - STEP 2: MODEL ARCHITECTURE
# Generator (ResNet) + Discriminator (PatchGAN)
# BUILDING BLOCK: ResNet BlocK

class ResNetBlock(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.block = nn.Sequential(
            nn.ReflectionPad2d(1),
            nn.Conv2d(channels, channels, 3),
            nn.InstanceNorm2d(channels),
            nn.ReLU(inplace=True),
            nn.ReflectionPad2d(1),
            nn.Conv2d(channels, channels, 3),
            nn.InstanceNorm2d(channels)
        )

    def forward(self, x):
        return x + self.block(x)   # Skip connection

# GENERATOR: ResNet-based (Encoder → ResBlocks → Decoder)

class Generator(nn.Module):
    """
    G_AB : Sketch → Photo
    G_BA : Photo  → Sketch
    Dono same architecture use karte hain
    """
    def __init__(self, in_ch=3, out_ch=3, n_res=6, ngf=64):
        super().__init__()

        #Encoder (downsampling)
        encoder = [
            nn.ReflectionPad2d(3),
            nn.Conv2d(in_ch, ngf, 7),
            nn.InstanceNorm2d(ngf),
            nn.ReLU(inplace=True),

            nn.Conv2d(ngf,     ngf*2, 3, stride=2, padding=1),
            nn.InstanceNorm2d(ngf*2),
            nn.ReLU(inplace=True),

            nn.Conv2d(ngf*2,   ngf*4, 3, stride=2, padding=1),
            nn.InstanceNorm2d(ngf*4),
            nn.ReLU(inplace=True),
        ]

        # ResNet Blocks (transformation)
        res_blocks = [ResNetBlock(ngf*4) for _ in range(n_res)]

        # Decoder (upsampling)
        decoder = [
            nn.ConvTranspose2d(ngf*4, ngf*2, 3, stride=2, padding=1, output_padding=1),
            nn.InstanceNorm2d(ngf*2),
            nn.ReLU(inplace=True),

            nn.ConvTranspose2d(ngf*2, ngf,   3, stride=2, padding=1, output_padding=1),
            nn.InstanceNorm2d(ngf),
            nn.ReLU(inplace=True),

            nn.ReflectionPad2d(3),
            nn.Conv2d(ngf, out_ch, 7),
            nn.Tanh()              # Output range [-1, 1]
        ]

        self.model = nn.Sequential(*encoder, *res_blocks, *decoder)

    def forward(self, x):
        return self.model(x)

# DISCRIMINATOR: PatchGAN

class Discriminator(nn.Module):
    """
    D_A : Sketch domain classificatin
    D_B : Photo  domain classification
    PatchGAN: not whole image but classify patches only
    """
    def __init__(self, in_ch=3, ndf=64):
        super().__init__()

        self.model = nn.Sequential(
            # Layer 1 - no norm
            nn.Conv2d(in_ch, ndf,   4, stride=2, padding=1),
            nn.LeakyReLU(0.2, inplace=True),

            # Layer 2
            nn.Conv2d(ndf,   ndf*2, 4, stride=2, padding=1),
            nn.InstanceNorm2d(ndf*2),
            nn.LeakyReLU(0.2, inplace=True),

            # Layer 3
            nn.Conv2d(ndf*2, ndf*4, 4, stride=2, padding=1),
            nn.InstanceNorm2d(ndf*4),
            nn.LeakyReLU(0.2, inplace=True),

            # Layer 4
            nn.Conv2d(ndf*4, ndf*8, 4, stride=1, padding=1),
            nn.InstanceNorm2d(ndf*8),
            nn.LeakyReLU(0.2, inplace=True),

            # Output patch map
            nn.Conv2d(ndf*8, 1,     4, stride=1, padding=1)
            # No sigmoid - BCEWithLogitsLoss use karenge
        )

    def forward(self, x):
        return self.model(x)


# MODELS INITIALIZE + DUAL GPU SETUP

def weights_init(m):
    """Conv aur BatchNorm weights initialize karo"""
    classname = m.__class__.__name__
    if 'Conv' in classname:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif 'BatchNorm' in classname or 'InstanceNorm' in classname:
        if m.weight is not None:
            nn.init.normal_(m.weight.data, 1.0, 0.02)
        if m.bias is not None:
            nn.init.constant_(m.bias.data, 0)

#Create models
G_AB = Generator(n_res=cfg.N_RES_BLOCKS).to(device)   # Sketch → Photo
G_BA = Generator(n_res=cfg.N_RES_BLOCKS).to(device)   # Photo  → Sketch
D_A  = Discriminator().to(device)                      # Sketch discriminator
D_B  = Discriminator().to(device)                      # Photo  discriminator

# Weights initialize
G_AB.apply(weights_init)
G_BA.apply(weights_init)
D_A.apply(weights_init)
D_B.apply(weights_init)

# Dual GPU (DataParallel) 
if torch.cuda.device_count() > 1:
    G_AB = nn.DataParallel(G_AB)
    G_BA = nn.DataParallel(G_BA)
    D_A  = nn.DataParallel(D_A)
    D_B  = nn.DataParallel(D_B)
    print(f"DataParallel enabled on {torch.cuda.device_count()} GPUs")


# LOSS FUNCTIONS + OPTIMIZERS


# Losses
criterion_GAN   = nn.MSELoss()         # Adversarial (LSGAN)
criterion_cycle = nn.L1Loss()          # Cycle consistency
criterion_id    = nn.L1Loss()          # Identity

# Optimizers
opt_G = torch.optim.Adam(
    list(G_AB.parameters()) + list(G_BA.parameters()),
    lr=cfg.LR, betas=cfg.BETAS
)
opt_D_A = torch.optim.Adam(D_A.parameters(), lr=cfg.LR, betas=cfg.BETAS)
opt_D_B = torch.optim.Adam(D_B.parameters(), lr=cfg.LR, betas=cfg.BETAS)

print("Optimizers ready!")


# MODEL SUMMARY


def count_params(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"\nModel Parameters:")
print(f"  G_AB (Sketch→Photo) : {count_params(G_AB):,}")
print(f"  G_BA (Photo→Sketch) : {count_params(G_BA):,}")
print(f"  D_A  (Sketch disc)  : {count_params(D_A):,}")
print(f"  D_B  (Photo  disc)  : {count_params(D_B):,}")


# QUICK FORWARD PASS TEST

G_AB.eval()
G_BA.eval()

with torch.no_grad():
    test_sketch = next(iter(sketch_loader)).to(device)
    test_photo  = next(iter(photo_loader)).to(device)

    fake_photo      = G_AB(test_sketch)     # Sketch → Photo
    recon_sketch    = G_BA(fake_photo)      # Photo  → Sketch (cycle)
    fake_sketch     = G_BA(test_photo)      # Photo  → Sketch
    recon_photo     = G_AB(fake_sketch)     # Sketch → Photo  (cycle)

    print(f"\nForward pass test:")
    print(f"  Input  sketch : {test_sketch.shape}")
    print(f"  Fake   photo  : {fake_photo.shape}")
    print(f"  Recon  sketch : {recon_sketch.shape}")
    print(f"  Input  photo  : {test_photo.shape}")
    print(f"  Fake   sketch : {fake_sketch.shape}")
    print(f"  Recon  photo  : {recon_photo.shape}")

G_AB.train()
G_BA.train()

print("\nStep 2 complete! Model architecture ready.")
```

```python
# CYCLEGAN TRAINING - SKETCHY DATASET

import os, random, time
from pathlib import Path
from tqdm import tqdm
import numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms

# CONFIGURATION

class Config:
    SKETCHY_ROOT = "/kaggle/working/sketchy/256x256"
    PHOTO_SUBDIR = "photo/tx_000100000000"
    SKETCH_SUBDIR = "sketch/tx_000000000000"
    
    IMG_SIZE = 128
    BATCH_SIZE = 6
    NUM_EPOCHS = 50
    LR = 0.0002
    BETAS = (0.5, 0.999)
    
    LAMBDA_CYCLE = 10
    LAMBDA_IDENTITY = 5
    N_RES_BLOCKS = 6
    MAX_SAMPLES = 2000  # Change to 12000 for full dataset
    
    CHECKPOINT_DIR = "/kaggle/working/checkpoints"

cfg = Config()
os.makedirs(cfg.CHECKPOINT_DIR, exist_ok=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device} | GPUs: {torch.cuda.device_count()}")

# DATASET


transform = transforms.Compose([
    transforms.Resize((cfg.IMG_SIZE, cfg.IMG_SIZE)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

class SketchyDataset(Dataset):
    def __init__(self, data_root, photo_subdir, sketch_subdir, transform=None, max_samples=2000):
        self.transform = transform
        photo_path = Path(data_root) / photo_subdir
        sketch_path = Path(data_root) / sketch_subdir
        
        self.photos = []
        self.sketches = []
        
        for cat in photo_path.iterdir():
            if cat.is_dir():
                self.photos.extend(list(cat.glob("*.jpg")))
        for cat in sketch_path.iterdir():
            if cat.is_dir():
                self.sketches.extend(list(cat.glob("*.png")))
        
        self.photos = self.photos[:max_samples]
        self.sketches = self.sketches[:max_samples]
        print(f"✅ Photos: {len(self.photos)} | Sketches: {len(self.sketches)}")
    
    def __len__(self):
        return min(len(self.photos), len(self.sketches))
    
    def __getitem__(self, idx):
        photo = Image.open(self.photos[idx % len(self.photos)]).convert('RGB')
        sketch = Image.open(self.sketches[idx % len(self.sketches)]).convert('RGB')
        if self.transform:
            photo = self.transform(photo)
            sketch = self.transform(sketch)
        return {'A': sketch, 'B': photo}

print("\n📦 Loading dataset...")
dataset = SketchyDataset(cfg.SKETCHY_ROOT, cfg.PHOTO_SUBDIR, cfg.SKETCH_SUBDIR, 
                         transform=transform, max_samples=cfg.MAX_SAMPLES)

sketches = [dataset[i]['A'] for i in range(len(dataset))]
photos = [dataset[i]['B'] for i in range(len(dataset))]

sketch_loader = DataLoader(sketches, batch_size=cfg.BATCH_SIZE, shuffle=True, drop_last=True)
photo_loader = DataLoader(photos, batch_size=cfg.BATCH_SIZE, shuffle=True, drop_last=True)

print(f"✅ Batches/epoch: {min(len(sketch_loader), len(photo_loader))}")

# MODELS

class ResidualBlock(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.block = nn.Sequential(
            nn.ReflectionPad2d(1), nn.Conv2d(channels, channels, 3),
            nn.InstanceNorm2d(channels), nn.ReLU(inplace=True),
            nn.ReflectionPad2d(1), nn.Conv2d(channels, channels, 3),
            nn.InstanceNorm2d(channels))
    def forward(self, x): return x + self.block(x)

class Generator(nn.Module):
    def __init__(self, in_channels=3, out_channels=3, n_res_blocks=6):
        super().__init__()
        model = [nn.ReflectionPad2d(3), nn.Conv2d(in_channels, 64, 7),
                 nn.InstanceNorm2d(64), nn.ReLU(inplace=True)]
        in_f, out_f = 64, 128
        for _ in range(2):
            model += [nn.Conv2d(in_f, out_f, 3, stride=2, padding=1),
                      nn.InstanceNorm2d(out_f), nn.ReLU(inplace=True)]
            in_f, out_f = out_f, out_f * 2
        for _ in range(n_res_blocks):
            model += [ResidualBlock(in_f)]
        out_f = in_f // 2
        for _ in range(2):
            model += [nn.ConvTranspose2d(in_f, out_f, 3, stride=2, padding=1, output_padding=1),
                      nn.InstanceNorm2d(out_f), nn.ReLU(inplace=True)]
            in_f, out_f = out_f, out_f // 2
        model += [nn.ReflectionPad2d(3), nn.Conv2d(64, out_channels, 7), nn.Tanh()]
        self.model = nn.Sequential(*model)
    def forward(self, x): return self.model(x)

class Discriminator(nn.Module):
    def __init__(self, in_channels=3):
        super().__init__()
        self.model = nn.Sequential(
            nn.Conv2d(in_channels, 64, 4, stride=2, padding=1), nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, 4, stride=2, padding=1), nn.InstanceNorm2d(128), nn.LeakyReLU(0.2),
            nn.Conv2d(128, 256, 4, stride=2, padding=1), nn.InstanceNorm2d(256), nn.LeakyReLU(0.2),
            nn.Conv2d(256, 512, 4, stride=1, padding=1), nn.InstanceNorm2d(512), nn.LeakyReLU(0.2),
            nn.Conv2d(512, 1, 4, stride=1, padding=1))
    def forward(self, x): return self.model(x)

# Initialize models
print("\n🏗️ Initializing models...")
G_S2P = Generator(3, 3, cfg.N_RES_BLOCKS).to(device)
G_P2S = Generator(3, 3, cfg.N_RES_BLOCKS).to(device)
D_P = Discriminator(3).to(device)
D_S = Discriminator(3).to(device)

def init_weights(m):
    if isinstance(m, (nn.Conv2d, nn.ConvTranspose2d)):
        nn.init.normal_(m.weight, 0.0, 0.02)
        if m.bias is not None: nn.init.constant_(m.bias, 0)
    elif isinstance(m, nn.InstanceNorm2d):
        if m.weight is not None: nn.init.normal_(m.weight, 1.0, 0.02)
        if m.bias is not None: nn.init.constant_(m.bias, 0)

G_S2P.apply(init_weights); G_P2S.apply(init_weights)
D_P.apply(init_weights); D_S.apply(init_weights)
print("✅ Models ready")

# Optimizers & Losses
opt_G = optim.Adam(list(G_S2P.parameters()) + list(G_P2S.parameters()), lr=cfg.LR, betas=cfg.BETAS)
opt_D_S = optim.Adam(D_S.parameters(), lr=cfg.LR, betas=cfg.BETAS)
opt_D_P = optim.Adam(D_P.parameters(), lr=cfg.LR, betas=cfg.BETAS)

criterion_GAN = nn.MSELoss()
criterion_cycle = nn.L1Loss()
criterion_id = nn.L1Loss()

# Replay Buffer
class ReplayBuffer:
    def __init__(self, max_size=50):
        self.max_size = max_size
        self.data = []
    def push_and_pop(self, data):
        result = []
        for element in data:
            element = element.unsqueeze(0)
            if len(self.data) < self.max_size:
                self.data.append(element)
                result.append(element)
            else:
                if random.random() > 0.5:
                    idx = random.randint(0, self.max_size - 1)
                    result.append(self.data[idx].clone())
                    self.data[idx] = element
                else:
                    result.append(element)
        return torch.cat(result)

buffer_S = ReplayBuffer()
buffer_P = ReplayBuffer()

# LR Scheduler
def get_scheduler(opt, n_epochs=50, decay_start=25):
    return torch.optim.lr_scheduler.LambdaLR(opt, lambda e: 1.0 if e < decay_start else 1.0 - (e - decay_start) / (n_epochs - decay_start))

sched_G = get_scheduler(opt_G)
sched_DS = get_scheduler(opt_D_S)
sched_DP = get_scheduler(opt_D_P)


# TRAINING LOOP WITH PROGRESS BAR

print("\n" + "="*60)
print("🚀 TRAINING STARTING...")
print("="*60)

total_batches = min(len(sketch_loader), len(photo_loader))

for epoch in range(cfg.NUM_EPOCHS):
    G_S2P.train(); G_P2S.train()
    D_S.train(); D_P.train()
    
    epoch_losses = {'G': 0, 'D_S': 0, 'D_P': 0, 'cycle': 0, 'id': 0}
    start_time = time.time()
    
    # Progress bar for batches
    pbar = tqdm(zip(sketch_loader, photo_loader), total=total_batches, 
                desc=f"Epoch {epoch+1}/{cfg.NUM_EPOCHS}", leave=True)
    
    for real_sketch, real_photo in pbar:
        real_sketch = real_sketch.to(device)
        real_photo = real_photo.to(device)
        batch_size = real_sketch.size(0)
        
        real_label = torch.ones(batch_size, 1, 14, 14).to(device) * 0.9
        fake_label = torch.zeros(batch_size, 1, 14, 14).to(device)
        
        # Generator
        opt_G.zero_grad()
        fake_photo = G_S2P(real_sketch)
        fake_sketch = G_P2S(real_photo)
        recon_sketch = G_P2S(fake_photo)
        recon_photo = G_S2P(fake_sketch)
        id_sketch = G_P2S(real_sketch)
        id_photo = G_S2P(real_photo)
        
        loss_G_adv = (criterion_GAN(D_P(fake_photo), real_label) + 
                      criterion_GAN(D_S(fake_sketch), real_label))
        loss_cycle = (criterion_cycle(recon_sketch, real_sketch) + 
                      criterion_cycle(recon_photo, real_photo)) * cfg.LAMBDA_CYCLE
        loss_id = (criterion_id(id_sketch, real_sketch) + 
                   criterion_id(id_photo, real_photo)) * cfg.LAMBDA_IDENTITY
        loss_G = loss_G_adv + loss_cycle + loss_id
        loss_G.backward()
        opt_G.step()
        
        # Discriminator Sketch
        opt_D_S.zero_grad()
        fake_sketch_buf = buffer_S.push_and_pop(fake_sketch.detach())
        loss_D_S = (criterion_GAN(D_S(real_sketch), real_label) + 
                    criterion_GAN(D_S(fake_sketch_buf), fake_label)) * 0.5
        loss_D_S.backward()
        opt_D_S.step()
        
        # Discriminator Photo
        opt_D_P.zero_grad()
        fake_photo_buf = buffer_P.push_and_pop(fake_photo.detach())
        loss_D_P = (criterion_GAN(D_P(real_photo), real_label) + 
                    criterion_GAN(D_P(fake_photo_buf), fake_label)) * 0.5
        loss_D_P.backward()
        opt_D_P.step()
        
        # Track losses
        epoch_losses['G'] += loss_G.item()
        epoch_losses['D_S'] += loss_D_S.item()
        epoch_losses['D_P'] += loss_D_P.item()
        epoch_losses['cycle'] += (loss_cycle.item() / 2)  # Average
        epoch_losses['id'] += (loss_id.item() / 2)
        
        # Update progress bar
        pbar.set_postfix({
            'G': f'{loss_G.item():.3f}',
            'D_S': f'{loss_D_S.item():.3f}',
            'D_P': f'{loss_D_P.item():.3f}'
        })
    
    # Average losses
    n = total_batches
    avg_G = epoch_losses['G'] / n
    avg_DS = epoch_losses['D_S'] / n
    avg_DP = epoch_losses['D_P'] / n
    avg_cycle = epoch_losses['cycle'] / n
    avg_id = epoch_losses['id'] / n
    
    epoch_time = time.time() - start_time
    print(f"📊 Epoch {epoch+1} | G:{avg_G:.3f} D_S:{avg_DS:.3f} D_P:{avg_DP:.3f} Cycle:{avg_cycle:.3f} | {epoch_time:.1f}s")
    
    # Update schedulers
    sched_G.step(); sched_DS.step(); sched_DP.step()
    
    # Save checkpoint every 5 epochs
    if (epoch + 1) % 5 == 0:
        torch.save({
            'epoch': epoch + 1,
            'G_S2P': G_S2P.state_dict(),
            'G_P2S': G_P2S.state_dict(),
            'D_S': D_S.state_dict(),
            'D_P': D_P.state_dict(),
            'opt_G': opt_G.state_dict(),
            'opt_D_S': opt_D_S.state_dict(),
            'opt_D_P': opt_D_P.state_dict(),
        }, f"{cfg.CHECKPOINT_DIR}/cyclegan_epoch_{epoch+1}.pth")
        print(f"💾 Checkpoint saved: epoch {epoch+1}")

# Save final model
torch.save({
    'G_S2P': G_S2P.state_dict(),
    'G_P2S': G_P2S.state_dict(),
    'D_S': D_S.state_dict(),
    'D_P': D_P.state_dict(),
}, f"{cfg.CHECKPOINT_DIR}/cyclegan_final.pth")

print("\n" + "="*60)
print("🎉 TRAINING COMPLETE!")
print(f"✅ Final model saved: {cfg.CHECKPOINT_DIR}/cyclegan_final.pth")
print("="*60)
```

```python
# LOCAL INFERENCE + EVALUATION (No Internet Required)

import torch
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr
from pathlib import Path

# Load model locally
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
checkpoint = torch.load('checkpoints/cyclegan_epoch_50.pth', map_location='cpu')

# Generator class (same as training)
class ResidualBlock(torch.nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.block = torch.nn.Sequential(
            torch.nn.ReflectionPad2d(1),
            torch.nn.Conv2d(channels, channels, 3),
            torch.nn.InstanceNorm2d(channels),
            torch.nn.ReLU(inplace=True),
            torch.nn.ReflectionPad2d(1),
            torch.nn.Conv2d(channels, channels, 3),
            torch.nn.InstanceNorm2d(channels)
        )
    def forward(self, x):
        return x + self.block(x)

class Generator(torch.nn.Module):
    def __init__(self, in_channels=3, out_channels=3, n_res_blocks=6):
        super().__init__()
        model = [
            torch.nn.ReflectionPad2d(3),
            torch.nn.Conv2d(in_channels, 64, 7),
            torch.nn.InstanceNorm2d(64),
            torch.nn.ReLU(inplace=True)
        ]
        in_f, out_f = 64, 128
        for _ in range(2):
            model += [torch.nn.Conv2d(in_f, out_f, 3, stride=2, padding=1),
                      torch.nn.InstanceNorm2d(out_f), torch.nn.ReLU(inplace=True)]
            in_f, out_f = out_f, out_f * 2
        for _ in range(n_res_blocks):
            model += [ResidualBlock(in_f)]
        out_f = in_f // 2
        for _ in range(2):
            model += [torch.nn.ConvTranspose2d(in_f, out_f, 3, stride=2, padding=1, output_padding=1),
                      torch.nn.InstanceNorm2d(out_f), torch.nn.ReLU(inplace=True)]
            in_f, out_f = out_f, out_f // 2
        model += [torch.nn.ReflectionPad2d(3), torch.nn.Conv2d(64, out_channels, 7), torch.nn.Tanh()]
        self.model = torch.nn.Sequential(*model)
    def forward(self, x):
        return self.model(x)

# Load weights
G = Generator()
G.load_state_dict(checkpoint['G_S2P'])
G.eval()
G = G.to(device)

print("✅ Model loaded locally!")

# Test with a sample sketch (agar hai toh)
import random
from pathlib import Path

# Agar sketches folder hai toh use karo
sketch_folder = Path('sketches')
if sketch_folder.exists():
    sketch_files = list(sketch_folder.glob('*.png')) + list(sketch_folder.glob('*.jpg'))
    if sketch_files:
        test_sketch = Image.open(random.choice(sketch_files)).convert('RGB')
        
        # Transform
        from torchvision import transforms
        transform = transforms.Compose([
            transforms.Resize((128, 128)),
            transforms.ToTensor(),
            transforms.Normalize([0.5]*3, [0.5]*3)
        ])
        
        input_tensor = transform(test_sketch).unsqueeze(0).to(device)
        
        # Generate
        with torch.no_grad():
            output = G(input_tensor)
        
        # Display
        def denorm(t):
            return (t * 0.5) + 0.5
        
        fig, axes = plt.subplots(1, 2, figsize=(10, 5))
        axes[0].imshow(test_sketch)
        axes[0].set_title('Input Sketch')
        axes[0].axis('off')
        
        out_img = denorm(output[0]).cpu().permute(1,2,0).numpy()
        axes[1].imshow(out_img)
        axes[1].set_title('Generated Photo')
        axes[1].axis('off')
        
        plt.show()
        print("✅ Inference successful!")
    else:
        print("No sketch images found in 'sketches' folder")
else:
    print("Create a 'sketches' folder and add some sketch images")
```
